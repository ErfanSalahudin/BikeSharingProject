## Data Wrangling

### Gathering Data

```python
day_df = pd.read_csv("day.csv")
day_df.head()
```

```python
hour_df = pd.read_csv("hour.csv")
hour_df.head()
```

## Exploratory Data Analysis (EDA)

```python
 hour_df.describe(include="all")
 day_df.describe(include="all")
```

### Explore ...

```python
import matplotlib.pyplot as plt

daily_rentals = day_df.groupby('dteday')['cnt'].sum().sort_values(ascending=False)

print(daily_rentals.head())
```

```python
import matplotlib.pyplot as plt

hourly_rentals = hour_df.groupby('hr')['cnt'].sum().sort_values(ascending=False)
print(hourly_rentals)
```

## Visualization & Explanatory Analysis

### Pada tanggal berapa penyewaan banyak terjadi?

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(12, 6))
daily_rentals.head(10).plot(kind='bar')
plt.title('Top 10 Days with Most Bike Rentals')
plt.xlabel('Date')
plt.ylabel('Total Rentals')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

### Pada jam berapa penyewaan banyak terjadi?

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(10, 6))
hourly_rentals.plot(kind='bar')
plt.title('Bike Rentals by Hour')
plt.xlabel('Hour of the Day')
plt.ylabel('Total Rentals')
plt.xticks(rotation=0)
plt.tight_layout()
plt.show()
```

- Conclusion dari pertanyaan ke 2 : Berdasarkan grafik diatas penyewaan paling banyak yaitu pada jam 17:00 yang memiliki total hampir mencapai 350000 penyewaan sedangkan penyewaan paling sedikit yaitu pada jam 04:00 yang memiliki total penyewa kurang dari 50000.
